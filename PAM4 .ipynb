**Why Modern Ethernet uses PAM4 over NRZ binary encoding streaming technique. **

Modern Ethernet has shifted towards PAM4 (Pulse Amplitude Modulation with 4 levels) over traditional NRZ (Non-Return to Zero) binary encoding for several reasons:

1. **Increased Data Rates**: PAM4 allows for higher data rates compared to NRZ. With PAM4, each symbol carries more bits of information, enabling higher throughput without requiring an increase in bandwidth.

2. **Efficiency**: PAM4 achieves higher spectral efficiency compared to NRZ. By transmitting multiple bits per symbol, PAM4 can achieve higher data rates within the same bandwidth.

3. **Spectral Efficiency**: PAM4 enables better spectral efficiency, which is crucial for accommodating increasing data demands within limited frequency spectra. With more bits per symbol, PAM4 can transmit more information in the same bandwidth, making it suitable for high-speed communication systems like modern Ethernet.

4. **Power Efficiency**: PAM4 can achieve better power efficiency compared to NRZ, especially at higher data rates. Transmitting multiple bits per symbol can reduce the energy per bit, contributing to overall power savings in Ethernet systems.

5. **Compatibility with Advanced Modulation Techniques**: PAM4 is compatible with advanced modulation techniques like coherent detection, which are essential for high-speed optical communication systems commonly used in data centers and high-performance computing environments. These techniques enhance signal-to-noise ratio and improve transmission reliability.

6. **Future-Proofing**: PAM4 offers scalability for future Ethernet standards and technologies. As data rates continue to increase, PAM4 provides a flexible framework for accommodating higher throughput requirements without requiring significant changes to the underlying physical layer technology.

Overall, the adoption of PAM4 in modern Ethernet reflects the need for higher data rates, improved spectral and power efficiency, compatibility with advanced modulation techniques, and future-proofing against evolving technological demands.


```python
import numpy as np
import matplotlib.pyplot as plt

def generate_NRZ_pulse(bitstream, amplitude):
    nrz_pulse = []
    for bit in bitstream:
        if bit == '0':
            pulse = np.zeros(100) - amplitude
        elif bit == '1':
            pulse = np.ones(100) * amplitude
        nrz_pulse.extend(pulse)
    return nrz_pulse

def generate_PAM4_modulation(bitstream):
    pam4_modulation = []
    for i in range(0, len(bitstream), 2):
        if bitstream[i:i+2] == '00':
            pulse = np.zeros(100)
        elif bitstream[i:i+2] == '01':
            pulse = np.ones(100)
        elif bitstream[i:i+2] == '10':
            pulse = np.ones(100) * 2
        elif bitstream[i:i+2] == '11':
            pulse = np.ones(100) * 3
        pam4_modulation.extend(pulse)
    return pam4_modulation

bitstream = "1010111011010011001101011"

# Generate NRZ square pulse
nrz_pulse = generate_NRZ_pulse(bitstream, 2)

# Plot NRZ square pulse
plt.subplot(2, 1, 1)
plt.plot(range(len(nrz_pulse)), nrz_pulse)
plt.title("NRZ Square Pulse")
plt.xlabel("Time")
plt.ylabel("Amplitude")

# Generate PAM4 modulation
pam4_modulation = generate_PAM4_modulation(bitstream)

# Plot PAM4 modulation
plt.subplot(2, 1, 2)
plt.plot(range(len(pam4_modulation)), pam4_modulation)
plt.title("PAM4 Modulation")
plt.xlabel("Time")
plt.ylabel("Amplitude")

plt.tight_layout()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_constellation_diagram(bitstream):
    constellation_points = []
    for i in range(0, len(bitstream), 2):
        if bitstream[i:i+2] == '00':
            constellation_points.append((0, 0))
        elif bitstream[i:i+2] == '01':
            constellation_points.append((0, 1))
        elif bitstream[i:i+2] == '10':
            constellation_points.append((1, 0))
        elif bitstream[i:i+2] == '11':
            constellation_points.append((1, 1))
    return constellation_points

bitstream = "1010111011010011001101011"

# Generate constellation diagram points
constellation_points = generate_constellation_diagram(bitstream)

# Plot constellation diagram
plt.figure(figsize=(6, 6))
for point in constellation_points:
    plt.plot(point[0], point[1], 'bo')
plt.title("PAM4 Modulation Constellation Diagram")
plt.xlabel("Real Part")
plt.ylabel("Imaginary Part")
plt.grid(True)
plt.xticks([0, 1])
plt.yticks([0, 1])
plt.xlim(-0.1, 1.1)
plt.ylim(-0.1, 1.1)
plt.show()
```
